```python
from birdnetlib import Recording
from birdnetlib.analyzer import Analyzer

analyzer = Analyzer()
recording = Recording(
    analyzer,
    "/teamspace/studios/opensoundscape/extended_audios_rtbc/32PC1_2021-22_0.wav",
)
recording.extract_embeddings()
print(recording.embeddings)
```

```python
import numpy as np

# Convertir embeddings a un array de NumPy
embeddings_array = np.array(recording.embeddings)

# Ver la forma del array
print(embeddings_array.shape)
print(embeddings_array[-1])  # Primer embedding

```

```python
import numpy as np

# Extraer solo los vectores de embedding de cada segmento
embeddings_array = np.array([segment['embeddings'] for segment in recording.embeddings])

# Ver la forma del array
print(embeddings_array.shape)  # (n_segmentos, embedding_dim)
```
